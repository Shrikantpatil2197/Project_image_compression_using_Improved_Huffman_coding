```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
from termcolor import colored
from PIL import Image as im 

input_path = input("Enter 512*512 image that is to be compressed:\n")
input_extension = input('Enter extention of image that is to be compressed:eg.jpg or.png or .jpeg\n')

img = cv2.imread(input_path + input_extension,0)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if (img[i][j] % 10 <= 5):
            img[i][j] = img[i][j] - img[i][j] % 10
        if (img[i][j] % 10 > 5):
            img[i][j] = img[i][j] + (10 - img[i][j] % 10)
print(colored('input_image pixel values:','red', attrs=['bold']))
print(img,'\n')
img1 = img[:128,:]
# print(img1)
# print(img1.shape)
img2 = img[128:256,:]
# print(img2)
# print(img2.shape)
img3 = img[256:384,:]
# print(img3)
# print(img3.shape)
img4 = img[384:512,:]
# print(img4)
# print(img4.shape)
image_all = [img1 ,img2, img3, img4]
try_img = []
for i in range(0,4):
    output_img = im.fromarray(image_all[i])
    output_img.save(input_path + f"{i}" + ".bmp")
    print(colored(f'input_image {i} size in bytes:','red', attrs=['bold']))
    print(os.stat(input_path + f"{i}" + ".bmp").st_size,'bytes\n')
    image = cv2.imread(input_path + f"{i}" + ".bmp",0)
    plt.imshow(image, cmap= 'gray')
    print(colored(f"\nInput image {i} that is to be compressed:",'red', attrs=['bold']))
    plt.show()
    if i==0:
        try_img = image_all[0]
    if (i<4) and i!=0 :
        try_img = np.vstack((try_img,image_all[i]))
plt.imshow(try_img, 'gray')
plt.show()
```
